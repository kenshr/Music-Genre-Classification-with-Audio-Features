```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

from tqdm import tqdm

import lyricsgenius as lg

%matplotlib inline
plt.style.use('ggplot')
```

```python
selected_genres = ['country','emo','gospel','hip-hop','rock']
lyrics_df = df[df.genre.isin(selected_genres)].reset_index(drop=True)
```

```python
lyrics_df
```

```python
dupe_check = lyrics_df.groupby(['track_name','artist_name']).size()
dupe_check[dupe_check > 1]
```

```python
lyrics_df.drop_duplicates(subset=['track_name','artist_name'],inplace=True)
```

```python
lyrics_df.genre.value_counts()
```

```python
def lyric_generator(track_lst,artist_lst):
    """
    Takes two ordered, same-size lists of tracks and artists and
    returns a list of each track's lyrics.
    
    Parameters:
    -----------
    track_lst: list of song titles
    artist_lst: list of artist names
    
    Output:
    -------
    Returns a list of lyrics (strings) for each song passed in.
    """
    lyric_lst = []
    # Initialize API object
    genius=lg.Genius(access_token='Insert access token',
                     skip_non_songs=True, 
                     remove_section_headers=True,
                     sleep_time=0.1,
                     verbose=False)
    
    # Iterate through tracks and store lyrics
    for t, a in tqdm(zip(track_lst, artist_lst)):
        try:
            song = genius.search_song(title = t,
                                      artist = a,
                                      get_full_info = False)
            if song is not None:
                lyric_lst.append(song.lyrics)
            else:
                lyric_lst.append('')
        except:
            lyric_lst.append('')
    return lyric_lst
```

```python
track_lst = lyrics_df.track_name.values.tolist()
artist_lst = lyrics_df.artist_name.values.tolist()

lyrics = lyric_generator(track_lst, artist_lst)
```

```python
l = pd.Series(lyrics)
l
```

```python
l.rename('lyrics',inplace=True)
```

```python
lyrics_df.reset_index(drop=True,inplace=True)
```

```python
lyrics_df
```

```python
final = pd.concat([lyrics_df, l],axis=1)
```
